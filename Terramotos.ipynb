Rango Intercuartílico

```python
iqr = LunasJupiter['mass_kg'].quantile(0.75) - LunasJupiter['mass_kg'].quantile(0.25)
filtro_inferior = LunasJupiter['mass_kg'] > LunasJupiter['mass_kg'].quantile(0.25) - (iqr * 1.5)
filtro_superior = LunasJupiter['mass_kg'] < LunasJupiter['mass_kg'].quantile(0.75) + (iqr * 1.5)

df_filtrado = LunasJupiter[filtro_inferior & filtro_superior]

sns.boxplot(df_filtrado['mass_kg'])
```
