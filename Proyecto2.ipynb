Rango Intercuartílico

```python
iqr = df['mass_kg'].quantile(0.75) - df['mass_kg'].quantile(0.25)
filtro_inferior = df['mass_kg'] > df['v'].quantile(0.25) - (iqr * 1.5)
filtro_superior = df['mass_kg'] < df['mass_kg'].quantile(0.75) + (iqr * 1.5)

df_filtrado = df[filtro_inferior & filtro_superior]

sns.boxplot(df_filtrado['mass_kg'])
```
