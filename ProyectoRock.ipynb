Rango Intercuartílico

```python
iqr = df['length'].quantile(0.75) - df['length'].quantile(0.25)
filtro_inferior = df['length'] > df['length'].quantile(0.25) - (iqr * 1.5)
filtro_superior = df['length'] < df['length'].quantile(0.75) + (iqr * 1.5)

df_filtrado = df[filtro_inferior & filtro_superior]

sns.boxplot(df_filtrado['length'])
```
