Rango Intercuart√≠lico

```python
iqr = df['LogMass'].quantile(0.75) - df['LogMass'].quantile(0.25)
filtro_inferior = df['LogMass'] > df['LogMass'].quantile(0.25) - (iqr * 1.5)
filtro_superior = df['LogMass'] < df['LogMass'].quantile(0.75) + (iqr * 1.5)

df_filtrado = df[filtro_inferior & filtro_superior]

sns.boxplot(df_filtrado['LogMass'])
```
